# Welcome To matplotlib

Now some functions does not work here as it is browser for it to work we add "%matplotlib inline"

**Here instead of import whole matplotlib library we will take important things from it as shown**

```python
%matplotlib inline
from matplotlib import pyplot as plt
```

If we want to plot line of specific coordinates we do following( where 1,2,3 are x-coordinates and having their y-coordinates respectives as 4,5,6) 

```python
plt.plot([1,2,3],[4,5,6])
plt.show()
```

If we want to including Heading and labels for axes then we do following 

```python
plt.plot([1,2,3],[4,5,6])
plt.title("Matplotlib chart")
plt.ylabel("This is Y")
plt.xlabel("This is X")
plt.show()
```

if we want to suppose 2 lines then we do following

```python
x = [1,5,4]
y = [23,45,89]
x2 = [3,7,3]
y2 = [5,8,9]
plt.plot(x,y)
plt.plot(x2,y2)
plt.show()
```

If we want to change the style of plot we can do following

```python
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')

x = [1,5,4]
y = [23,45,89]
x2 = [3,7,3]
y2 = [5,8,9]
plt.plot(x,y)
plt.plot(x2,y2)
plt.show()
```

If we want to see heading(label) for each plot, then do following

```python
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')

x = [1,5,4]
y = [23,45,89]
x2 = [3,7,3]
y2 = [5,8,9]
plt.plot(x,y,label='first')
plt.plot(x2,y2,label='second')
plt.legend()
plt.show()
```

If we want specify width of line as well then we do following

```python
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')

x = [1,5,4]
y = [23,45,89]
x2 = [3,7,3]
y2 = [5,8,9]
plt.plot(x,y,label='first',linewidth=3)
plt.plot(x2,y2,label='second',linewidth=6)
plt.legend()
plt.show()
```

If we want one plot as scatter plot then we do

```python
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')

x = [1,5,4]
y = [23,45,89]
x2 = [3,7,3]
y2 = [5,8,9]
plt.scatter(x,y,label='first',linewidth=6)
plt.plot(x2,y2,label='second',linewidth=3)
plt.legend()
plt.show()
```

If we want to make histogram then we do following

```python
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')

x = [1,5,4]
plt.hist(x,label='first',linewidth=6)
plt.legend()
plt.show()
```
